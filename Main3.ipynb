```python
drop = ['RowNumber','CustomerId','Surname']
data.drop(drop,inplace=True, axis=1)
data
```

```python
data['Gender'] = data['Gender'].replace(("Male"),1)
data['Gender'] = data['Gender'].replace(("Female"),0)
```
